```python
#function for checking if relation defined on a set is correct or not
def check_relation(R,A):
    for i in range (0,len(R)):
        
        if R[i][0] not in A:
            print("Invalid Relation:",R[i][0],"is not an element of set A=",A)
            return False
        
        elif R[i][1] not in A:
            print("Invalid Relation:",R[i][1],"is not an element of set A=",A)
            return False
        
        else:
            continue
            
    return True

#function to check for reflexive property
def reflexive(A,R): 
    for i in A:
        a=(i,i)  #making a tuple (a,a) such that a belongs to set A
       
        if a not in R:
            print("Not Reflexive\nTuple Voilating the Property")
            return a
        
        else:
            continue
            
    a="Reflexive"
    return a

#function to check for symmetric property
def symetric(R):
    for i in range(0,len(R)):
        a=(R[i][1],R[i][0]) #for every element (a,b) in R, making a tuple a=(b,a)
        
        if a not in R:
            print("Not Symmetric\nTuple Voilating the Property:")
            a=(R[i][0],R[i][1])
            return a
        
        else:
            continue
            
    a="Symmetric"
    return a

#function to check for antisymmetric property
def antisymetric(R):
    for i in range(0,len(R)):
        a=(R[i][1],R[i][0])  #for every element (a,b) in R, making a tuple a=(b,a)
        
        if a in R and a[0]!=a[1]:   #if (a,b) and (b,a) in R and a!=b
            print("Not Antisymmetric\nTuple Voilating the Property:")
            a=(R[i][0],R[i][1])
            return a
        
        else:
            continue
            
    a="Antisymmetric"
    return a

#main
n=""
while (n!="n"):
    A=input("Enter the set A:\nA=")
    A=eval(A)
    Relation=input("Enter the relation R:\nR")
    Relation=eval(Relation)
    
    R=[]
    for i in Relation:
        R.append(i)
        
    if(check_relation(R,A)):
        print("\n---Checking Properties---")
        print(reflexive(A,R),"\n")
        print(antisymetric(R),"\n")
        print(symetric(R),"\n")
        
    n=input("Do you want to continue? y/n?")
    print("\n")
```
