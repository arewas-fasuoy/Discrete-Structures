```python
#function for swaping values
def swap(a,b):
    c=a
    a=b
    b=c
    return (a,b)

#function for taking negation
def negation(val):
    if val[0] == '~':
        return val[1:]
    else:
        return '~'+val
    
#function to find implication of proposition
def implification(form,value):
    left,right = value.split('->')
    if form == 'implification':
        return value
    elif form == 'contrapositive':
        left,right=swap(negation(left),negation(right))
        return (left+'->'+right)
    elif form == 'converse':
        left,right = swap(left,right)
        return (left+'->'+right)
    elif form == 'inverse':
        left,right = negation(left),negation(right)
        return (left+'->'+right)

#function for finding converse of proposition
def converse(value):
    left,right = value.split('->')
    left,right = swap(left,right)
    return (left+'->'+right)

#function for finding contrapositive of proposition
def contrapositive(value):
    left,right = value.split('->')
    left,right=swap(negation(left),negation(right))
    return (left+'->'+right)

#function for finding inverse of proposition
def inverse(value):
    left,right = value.split('->')
    left,right= negation(left),negation(right)
    return (left+'->'+right)

print('Enter the Proposition and its Form \n')
Form=input('Form')
Value=input("Value: ")
#function calls
temp=implification(Form,Value)
Converse=converse(temp)
Inverse=inverse(temp)
Contrapositive= contrapositive(temp)

print("Input:\n", Value, 'in', Form)
print("Output:\n")
print("Implification", temp)
print("Converse:", Converse)
print("Inverse:", Inverse)
print("Cotrapositive:", Contrapositive)    
```
